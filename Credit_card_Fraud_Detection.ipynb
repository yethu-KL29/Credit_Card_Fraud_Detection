```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
df = pd.read_csv("C:\\Users\\yethu.krishnan\\Desktop\\MY_PROJECT_SOURCES\\creditcard.csv")
```

```python
df
```

```python
#checking null value counts
df.isnull().sum()
```

```python
#checking  how many counts for each class
df['Class'].value_counts()
```

```python
legit  = df[df['Class']==0]
fraud = df[df['Class']==1]
```

```python
fraud
```

```python
#understanding the differnce of the data by taking by mean 
df.groupby('Class').mean()
```

```python
#dealing with unbalanced data by undersampling
legit_sample = legit.sample(n=492)
```

```python
legit_sample.shape
```

```python
#Cncatinating the 2 dataframes
new_df = pd.concat([legit_sample,fraud],axis=0)
```

```python
new_df
```

```python
new_df.groupby('Class').mean()
```

```python
#extracting the feautures into X and Y
X = new_df.drop(columns='Class',axis=1)
```

```python
Y = new_df['Class']
```

```python
X
```

```python
Y
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
X_train
```

```python
Y_train
```

```python
Y_train.value_counts()
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
#import The  Model 
model = LogisticRegression()
```

```python
# training the Logistic Regression Model with Training Data
model.fit(X_train, Y_train)
```

```python
# accuracy on training data
from sklearn.metrics import accuracy_score

X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy on Training data : ', training_data_accuracy)
```

```python
# accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
X_test_prediction
```

```python
print('Accuracy score on Test Data : ', test_data_accuracy)
```

```python
#checking outliers 
from scipy import stats

# Compute residuals
residuals = Y_train - X_train_prediction

# Compute z-scores
z_scores = (residuals - residuals.mean()) / residuals.std()



```

```python
z_scores
```

```python
threshold = 3
outliers = np.abs(z_scores) > threshold
```

```python
outliers.index = Y_train.index
outliers
```

```python
# Drop rows with outliers from both X_train and Y_train
new_X_train = X_train[~outliers]
new_Y_train = Y_train[~outliers]
```

```python
new_X_train
```

```python
new_Y_train
```

```python
model.fit(new_X_train, new_Y_train)
```

```python
new_X_train_prediction = model.predict(new_X_train)
training_data_accuracy = accuracy_score(new_X_train_prediction, new_Y_train)
```

```python
print('Accuracy on Training data : ', training_data_accuracy)
```

```python
# accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy on Training data : ', test_data_accuracy)
```
